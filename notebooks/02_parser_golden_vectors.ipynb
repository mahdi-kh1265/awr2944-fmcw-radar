# Parser Golden Vectors

This notebook demonstrates how to load synthetic golden vectors directly.

```python
from awr2944_dca.formats.synthetic import generate_golden_vector
from awr2944_dca.formats.layouts import awr2944_real_2lane_interleaved_candidate

# Create synthetic golden vector for a specific layout
samples = 256
chirps = 128
frames = 1

try:
    vec = generate_golden_vector(
        layout=awr2944_real_2lane_interleaved_candidate,
        samples_per_chirp=samples,
        chirps_per_frame=chirps,
        num_frames=frames
    )
    print(f"Generated vector of length {len(vec)}")
except NotImplementedError:
    print("Synthetic generation not yet implemented for all layouts.")
```
