# Spec Validation Demo

This notebook demonstrates how the validation logic pulls constraints directly from the hardware specification YAMLs.

```python
from pathlib import Path
from awr2944_dca.api.experiment import Experiment
from awr2944_dca.config.validation import Severity

exp = Experiment.init(name="notebook_spec_demo", preset="first-capture")
cfg = exp.get_config()

# Intentionally break a hardware spec
cfg.profile.start_freq_ghz = 82.0  # AWR2944 only supports 76-81 GHz

# Validate
from awr2944_dca.config.validation import validate_config
results = validate_config(cfg)

for r in results:
    if r.severity == Severity.ERROR:
        print(f"CAUGHT ERROR in {r.field}: {r.message}")
```
