# mmWave Studio Probe

This notebook demonstrates how to invoke the TI mmWave Studio offline probe from the Python API.

```python
import json
from awr2944_dca.api.experiment import Experiment

exp = Experiment.init(name="notebook_probe_demo", preset="first-capture")

# Find installations
installs = exp.find_studio()
print(f"Found {len(installs)} TI installations.")

# Generate probe script
probe_lua = exp.generate_ti_probe()
print(f"Generated Lua script at: {probe_lua}")
```

At this point, you would manually run the Lua script in the mmWave Studio Lua shell. After it completes, a `probe_result.json` file is written, which we can parse.

```python
# Simulate the Lua script running and writing the JSON
result_json = exp.root_dir / "ti" / "probe_logs" / "probe_result.json"
result_json.write_text('{"probe_executed": true, "ar1_available": false, "write_to_log_available": true}')

# Parse results
data = json.loads(result_json.read_text())
print("Probe Results:")
for k, v in data.items():
    print(f"{k}: {v}")
```
