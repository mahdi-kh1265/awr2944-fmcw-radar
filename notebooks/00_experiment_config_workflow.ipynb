# Experiment Config Workflow

This notebook demonstrates how to scaffold and interact with experiments programmatically using the AWR2944 DCA Lab API.

```python
from pathlib import Path
from awr2944_dca.api.experiment import Experiment

# Scaffold a new experiment
exp = Experiment.init(name="notebook_test_experiment", preset="first-capture")

print(f"Experiment initialized at: {exp.root_dir}")
```

```python
# Set devices
exp.set_device("AWR2944EVM")
exp.set_capture("DCA1000EVM")

# Check config validity
results = exp.check()
for r in results:
    print(f"{r.field}: {r.severity.name} - {r.message}")
```

```python
# Get summary
summary = exp.summary()
print(f"Expected file size: {summary['expected_bytes'] / (1024*1024):.2f} MiB")
```
